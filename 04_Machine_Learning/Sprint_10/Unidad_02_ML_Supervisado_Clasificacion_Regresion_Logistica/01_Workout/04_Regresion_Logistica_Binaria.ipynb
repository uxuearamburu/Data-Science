### Entender el problema

Nos piden un modelo que prediga si los clientes van a pagar o impagar para poder hacer acciones preventivas en algn caso (clientes ya existentes) o comerciales (crditos preconcedidos). 

Parace muy claro, pero qu significa impagar? Es que maana dejo de pagar una factura?Y si la repongo en menos de 30 das?Es tener una deuda a los seis meses sin pagar? Esto tenemos que aclararlo, no esperar a deducirlo del datatset (Aunque en estos ejemplos lo hagamos as). En nuestro caso nos contestan que es tener deudas pendientes de ms de 2 aos. Perfecto. Vamos con ello.

### Prepaaracin Datos: Primer Vistazo

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data/cs-training.csv')

df.head()
```

### Preparacin Datos: Limpieza de datos

```python
df.rename(columns = {'SeriousDlqin2yrs': 'target'}, inplace=True)
```

```python
sns.countplot(data=df, x='target');
```

```python
df.info()
```

```python
# Nos cargamos columnas inutiles
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# Nos cargamos duplicados
df.dropna(inplace=True)
```

### Preparacin datos: Dividimos el dataset

```python
from sklearn.model_selection import train_test_split 

X = df.drop(['target'], axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=42) 
```

### MiniEDA: Matriz de correlacin
Vamos a cargarnos algunas columnas que no estn muy relacionadas con el target (seria ms aconsejable descartarlo haciendo un bivariante con test de hiptesis o usando pairplot, aunque hay que tener en cuenta que debe hacerse por partes porque pairplot es muy demandante y este dataset es grande en dato).

```python
df_eda = X_train.copy()
df_eda["target"] = y_train.copy()
```

```python
np.abs(df_eda.corr()['target']).sort_values(ascending=False)
```

```python
corr = np.abs(df_eda.corr()['target']).sort_values(ascending=False)

# Features con menos de 0.1 de correlacin vs el target
bad_corr_feat = corr[corr < 0.05].index.values

# Filtramos las columnas buenas
df_eda.drop(columns=bad_corr_feat, inplace=True)
```

```python
features_num = [col for col in df_eda.columns if "target" not in col]
X_train = X_train[features_num].copy()
```

### Modelo

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter = 5000)

log_reg.fit(X_train, y_train)
```

Veremos los hiperparmetros (en este caso argumentos al constructor de la clase LogisticRegression) a medida que vayamos usando el modelo. Ahora tratemos "max_iter":

`max_iter` = numero de iteraciones del optimizador (recuerda que el gradiente descendente va probando combinaciones hasta que encuentra una que no mejora la funcin de minimizacin despus de varios intentos), esta es otra forma de decirle el mximo nmero de combinaciones, de veces que se aplica el gradiennte descendente en caso de que no se llegue a la condicin anterior (denominada tolerancia)



```python
print(log_reg.coef_)
print(log_reg.intercept_)
print(log_reg.classes_)
```

### Feature importance

```python
intercept = log_reg.intercept_
coefs = log_reg.coef_.ravel()

features = pd.DataFrame(coefs, X_train.columns, columns=['coefficient']).copy()
features['coefficient'] = np.abs(features['coefficient'])

features.sort_values('coefficient', ascending=False).head()
```

```python
stdevs = []
for i in X_train.columns:
    stdev = df[i].std()
    stdevs.append(stdev)

features["stdev"] = np.array(stdevs).reshape(-1,1)
features["importance"] = features["coefficient"] * features["stdev"]
features['importance_standarized'] = features['importance'] / y_train.std()

features = features.sort_values('importance_standarized', ascending=True)
plt.barh(features.index, features.importance_standarized);
```

## Accuracy

Antes de usar el X_test hay que hacerles las mismas transformaciones que le hayamos hecho al X_train

```python
X_test = X_test[X_train.columns].copy()
```

```python
from sklearn.metrics import accuracy_score

acc_train = round(accuracy_score(log_reg.predict(X_train), y_train), 3)
acc_test = round(accuracy_score(log_reg.predict(X_test), y_test), 3)

print("Accuracy train:", acc_train)
print("Accuracy test:", acc_test)
```

### Confusion matrix

```python
from sklearn.metrics import confusion_matrix

c_matrix = confusion_matrix(y_test, log_reg.predict(X_test))

c_matrix_df = pd.DataFrame(c_matrix, columns = ['Pred_Pays', 'Pred_Default'],
                          index = ['True_Pays', 'True_Default'])

sns.heatmap(c_matrix_df, annot=True, fmt='g');
```

```python
sns.heatmap(c_matrix_df/np.sum(c_matrix_df), annot=True, 
            fmt='.2%', cmap='Blues');
```

## Classification report

```python
from sklearn.metrics import classification_report
from pprint import pprint

print(classification_report(y_test, log_reg.predict(X_test)))
```

## ROC Curve

```python
### Usando matplotlib
from sklearn.metrics import roc_curve, auc
scores = log_reg.predict_proba(X_test)
fpr, tpr, thresholds = roc_curve(y_test, scores[:,1])
roc_auc = auc(fpr,tpr)
print("AUROC: %.2f" %(roc_auc))
plt.figure(figsize=(6, 5))  
plt.plot(fpr, tpr, linewidth=2, color= "blue", label=f"ROC Logistic Reg (area = {roc_auc:0.2f})")
plt.plot([0, 1], [0, 1], 'k:', label="Random classifier's ROC curve")
plt.ylabel('True Positive Rate (Recall)')
plt.grid()
plt.axis([0, 1, 0, 1])
plt.legend(loc="lower right", fontsize=13)
```
